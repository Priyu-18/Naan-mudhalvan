```python
import pandas as pd
import seaborn as sns
data = pd.read_csv('House Price India.csv')
print(data.head())
```

univariate analysis on Numerical variables using the histogram function.

```python
sns.histplot(data['number of bedrooms'])
```

BI-VARIATE ANALYSIS

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
sns.barplot(x=data['living area'], y=data['number of floors'])

```

Numerical data


```python
sns.scatterplot(x=data['living area'],
				y=data['number of floors'])
```

Multi-variate analysis

```python
from sklearn import datasets, decomposition
iris = datasets.load_iris()
X = iris.data
y = iris.target
pca = decomposition.PCA(n_components=2)
X = pca.fit_transform(X)
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)
```
